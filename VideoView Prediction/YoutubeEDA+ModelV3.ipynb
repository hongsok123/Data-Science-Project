```python
lr = LinearRegression()
```

```python
data = pd.get_dummies(df)
```

```python
features =data.drop('vid_view', axis = 1)
target  =data.vid_view
```

```python
features.shape, target.shape
```

```python
model = lr.fit(features, target)

predictions = lr.predict(features)
```

```python
residule = target - predictions
plt.figure(figsize = (15,10))
sns.displot(residule,bins = 150, kde = True, height = 5, aspect = 2.5, color = 'green')
plt.xlabel('Residuals', fontsize  = 15)
plt.ylabel('Frequency', fontsize = 15)
plt.title('Distribution of Error Terms', fontsize = 25)
plt.show()
```

```python
mse = mean_squared_error(predictions,target)
rsquared = r2_score(target, predictions)
mae = mean_absolute_error(target, predictions)

print(f"Mean Absolute Error : {mae}, r2Sccore : {rsquared}")
```

```python
MAPE = mean_absolute_percentage_error(predictions, target)
MAPE
```

```python
score = 1-MAPE
print("Score = "+"{: .2%}".format(score))
```
